```python
#importing all required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error , r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('delaney.csv')
df.head()
```

***It is dataset of finding solubility or chemistry model to predict weather the chemical will be soluble or insoluble.***

```python
df
```

```python
#data preprocessing and cleaning the data
#seperating data as x and y
y = df['logS']
x = df.drop('logS',axis=1)#if axis = 1 it drop column else or 0 : drops row
x.head()
```

```python
#since we got both x and y
#splitting data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
#randomness for training and testing data
```

**Model Building**
** **
**Linear Regression**


```python
#training the model
lr = LinearRegression()
lr.fit(x_train, y_train)
```

**Applying model to make a Prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
#print(y_lr_test_pred,y_lr_train_preda
Data =  np.array(['y_lr_train_pred','y_lr_test_pred'])
```

```python
#Normalizing the data given|
max_y_test = np.max(y_test)
min_y_test =np.min(y_test)
max_y_train =np.max(y_train)
min_y_train =np.min(y_train)

normalized_y_train =(y_train- min_y_train)/(max_y_train-min_y_train)
normalized_y_test =(y_test- min_y_test)/(max_y_test-min_y_test)

#normalizing predicted data
min_y_lr_train_pred = np.min(y_lr_train_pred)
min_y_lr_test_pred = np.min(y_lr_test_pred)
max_y_lr_train_pred = np.max(y_lr_train_pred)
max_y_lr_test_pred = np.max(y_lr_test_pred)
normalized_y_train_pred = (y_lr_train_pred- min_y_lr_train_pred)/(max_y_lr_train_pred- min_y_lr_train_pred)
normalized_y_test_pred = (y_lr_test_pred- min_y_lr_test_pred)/(max_y_lr_test_pred- min_y_lr_test_pred)
```

```python
#printing normalized data
normalized_y_test.info()
normalized_y_test_pred
```

**Evaluation of Model Prediction**

```python
lr_train_mse = mean_squared_error(normalized_y_train, normalized_y_train_pred)
lr_train_r2 = r2_score(normalized_y_train,normalized_y_train_pred)

lr_test_mse = mean_squared_error(normalized_y_test, normalized_y_test_pred)
lr_test_r2 = r2_score(normalized_y_test,normalized_y_test_pred)
```

```python
print('lr_train_mse :', lr_train_mse)
print('lr_test_mse :', lr_test_mse)
print('lr_train_r2 :', lr_train_r2)
print('lr_test_r2 :', lr_test_r2)
```

Let's break down the results:

MSE (Mean Squared Error):

Train MSE: 0.0095
Test MSE: 0.0128
Lower MSE values are generally better as they indicate that the model's predictions are closer to the actual values. The fact that the training MSE is slightly lower than the test MSE suggests that your model is performing well on the training data and reasonably well on unseen data.
RÂ² (Coefficient of Determination):

Train RÂ²: 0.7348
Test RÂ²: 0.5724
RÂ² values range from 0 to 1, with 1 indicating a perfect fit. A training RÂ² of 0.7348 indicates that about 73.48% of the variance in the training data is explained by the model. The test RÂ² of 0.5724 means that 57.24% of the variance in the test data is explained by the model.


```python
lr_results = pd.DataFrame(['Linear Regression',[lr_train_mse],lr_test_mse,lr_test_r2,lr_train_r2]).transpose()
lr_results.columns = ['Method','Training MSE','Testing MSE','Testing r2','Training r2']
print(lr_results)
```
